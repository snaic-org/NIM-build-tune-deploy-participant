# Example: Using OpenAI API from Python
This notebook shows a minimal end‑to‑end call to the OpenAI Chat Completions API, and how we can change them to NVIDIA NIM API.

### Import dependencies

```python
import os
from openai import OpenAI
```

### Import API Keys

```python
from getpass import getpass
import os

# Prompt once per session; the input field will be blank as you type
os.environ["OPENAI_API_KEY"] = getpass("Enter your OpenAI API key: ")
```

```python
from getpass import getpass
import os

# Prompt once per session; the input field stays blank while you type
os.environ["NVIDIA_API_KEY"] = getpass("Enter your NVIDIA NIM API key: ")
```

```python
# Get OpenAI API key from environment
openai_api_key = os.getenv("OPENAI_API_KEY")

# Get NVIDIA API key from environment
nvidia_api_key = os.getenv("NVIDIA_API_KEY")
```

### Run Inference

```python
client = OpenAI(
    api_key=openai_api_key,
    base_url="https://api.openai.com/v1"
)

response = client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=[{"role": "user", "content": "What's your model name?"}],
)
print(response.choices[0].message.content)
```
